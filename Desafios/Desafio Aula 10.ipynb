Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias


```python
#importar a biblioteca pandas
import pandas as pd 
```

```python
#leitura do arquivo
consoles = pd.read_excel('best-selling game consoles.xlsx', sheet_name='consoles')
consoles.head(5)
```

```python
#contagem numero de consoles - groupby e agg
empresas = consoles.groupby('Company')['Console Name'].count()
empresas
```

Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente. 

```python
#contagem e criação de coluna para a contagem
empresas = consoles.groupby('Company')['Console Name'].count().reset_index()
#criar uma nova coluna chamada de 'Console Count' para diferenciar de 'Console Name'
empresas.rename(columns={'Console Name': 'Console Count'}, inplace=True)
#ordena em ordem decrescente pela nova coluna 'Console Count'
ordem_desc = empresas.sort_values(by='Console Count', ascending=False)
ordem_desc

```

Sobre o dataframe anterior, adicione a coluna "Market Value" presente no tabela do arquivo "company.txt".

```python
#leitura da base company.txt
market_value = pd.read_csv("company.txt")
market_value
```

```python
#juntando as bases com merge
console_mv = pd.merge(ordem_desc, market_value, on='Company', how='left')
console_mv
```

Qual o valor médio de Market Value por console?

```python
#soma dos valores em market value divididos pela some do número de consoles
media = console_mv['Market Value'].sum() / console_mv['Console Count'].sum()
media
```
